```
using System;
using System.Collections.Generic;

public delegate void TransactionHandler(string message);

public interface IInsurable
{
    void CheckInsuranceStatus();
}

public interface IRewardable
{
    void RedeemRewards(int amount);
}

public abstract class CreditCard
{
    private string cardNumber;
    private string holderName;
    private string expiryDate;
    private decimal balance;
    private decimal creditLimit;
    private decimal annualFee;
    private bool isActive;
    private List<string> transactionHistory = new List<string>();

    public event TransactionHandler OnTransaction;

    public string CardNumber
    {
        get { return cardNumber; }
        set { cardNumber = value; }
    }

    public string HolderName
    {
        get { return holderName; }
        set { holderName = value; }
    }

    public string ExpiryDate
    {
        get { return expiryDate; }
        set { expiryDate = value; }
    }

    public decimal Balance
    {
        get { return balance; }
        set { balance = value; }
    }

    public decimal CreditLimit
    {
        get { return creditLimit; }
        set { creditLimit = value; }
    }

    public decimal AnnualFee
    {
        get { return annualFee; }
        set { annualFee = value; }
    }

    public bool IsActive
    {
        get { return isActive; }
        set { isActive = value; }
    }

    public CreditCard(string _CardNumber, string _HolderName, string _ExpiryDate, decimal _Balance, decimal _CreditLimit, decimal _AnnualFee)
    {
        CardNumber = _CardNumber;
        HolderName = _HolderName;
        ExpiryDate = _ExpiryDate;
        Balance = _Balance;
        CreditLimit = _CreditLimit;
        AnnualFee = _AnnualFee;
        IsActive = true;
    }

    public virtual string GetInfo()
    {
        return $"Карта: {CardNumber}, Держатель: {HolderName}, Срок: {ExpiryDate}, Баланс: {Balance}, Лимит: {CreditLimit}, Годовой сбор: {AnnualFee}";
    }

    public virtual void Pay(decimal sum)
    {
        if (sum > Balance + CreditLimit)
        {
            TriggerTransactionEvent("Недостаточно средств для выполнения транзакции.");
        }
        else if (!IsActive)
        {
            TriggerTransactionEvent("Карта неактивна.");
        }
        else
        {
            if (sum > Balance)
            {
                CreditLimit -= (sum - Balance);
                Balance = 0;
            }
            else
            {
                Balance -= sum;
            }
            transactionHistory.Add($"Платеж: {sum}");
            TriggerTransactionEvent($"Транзакция успешна: {sum}. Баланс: {Balance}, Лимит: {CreditLimit}");
        }
    }

    public void ViewTransactionHistory()
    {
        Console.WriteLine("История транзакций:");
        foreach (var transaction in transactionHistory)
        {
            Console.WriteLine(transaction);
        }
    }

    public void ApplyAnnualFee()
    {
        if (Balance >= AnnualFee)
        {
            Balance -= AnnualFee;
            TriggerTransactionEvent($"Списан годовой сбор: {AnnualFee}. Баланс: {Balance}");
        }
        else
        {
            TriggerTransactionEvent("Недостаточно средств для списания годового сбора.");
        }
    }

    protected void TriggerTransactionEvent(string message)
    {
        OnTransaction?.Invoke(message);
    }
}

public class GoldCreditCard : CreditCard, IInsurable, IRewardable
{
    private decimal bonusMiles;
    private decimal insuranceCoverage;
    private List<string> rewardPrograms = new List<string>();

    public decimal BonusMiles
    {
        get { return bonusMiles; }
        set { bonusMiles = value; }
    }

    public decimal InsuranceCoverage
    {
        get { return insuranceCoverage; }
        set { insuranceCoverage = value; }
    }

    public GoldCreditCard(string _CardNumber, string _HolderName, string _ExpiryDate, decimal _Balance, decimal _CreditLimit, decimal _AnnualFee, decimal _BonusMiles, decimal _InsuranceCoverage)
        : base(_CardNumber, _HolderName, _ExpiryDate, _Balance, _CreditLimit, _AnnualFee)
    {
        BonusMiles = _BonusMiles;
        InsuranceCoverage = _InsuranceCoverage;
    }

    public override void Pay(decimal sum)
    {
        base.Pay(sum);
        int earnedMiles = (int)(sum / 10);
        BonusMiles += earnedMiles;
        TriggerTransactionEvent($"Начислено {earnedMiles} миль. Всего миль: {BonusMiles}");
    }

    public void AddRewardProgram(string program)
    {
        rewardPrograms.Add(program);
        Console.WriteLine($"Добавлена программа лояльности: {program}");
    }

    public void CheckInsuranceStatus()
    {
        Console.WriteLine($"Страховое покрытие: {InsuranceCoverage}");
    }

    public void RedeemRewards(int amount)
    {
        if (amount > BonusMiles)
        {
            TriggerTransactionEvent("Недостаточно миль.");
        }
        else
        {
            BonusMiles -= amount;
            TriggerTransactionEvent($"Списано {amount} миль. Остаток: {BonusMiles}");
        }
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Мили: {BonusMiles}, Страховка: {InsuranceCoverage}, Программы: {string.Join(", ", rewardPrograms)}";
    }
}

public class PremiumCreditCard : CreditCard
{
    private decimal cashbackAccumulated;
    private string supportAssistant;
    private int freeLoungeVisits;

    public decimal CashbackAccumulated
    {
        get { return cashbackAccumulated; }
        set { cashbackAccumulated = value; }
    }

    public string SupportAssistant
    {
        get { return supportAssistant; }
        set { supportAssistant = value; }
    }

    public int FreeLoungeVisits
    {
        get { return freeLoungeVisits; }
        set { freeLoungeVisits = value; }
    }

    public PremiumCreditCard(string _CardNumber, string _HolderName, string _ExpiryDate, decimal _Balance, decimal _CreditLimit, decimal _AnnualFee, string _SupportAssistant, int _FreeLoungeVisits)
        : base(_CardNumber, _HolderName, _ExpiryDate, _Balance, _CreditLimit, _AnnualFee)
    {
        SupportAssistant = _SupportAssistant;
        FreeLoungeVisits = _FreeLoungeVisits;
    }

    public override void Pay(decimal sum)
    {
        base.Pay(sum);
        decimal cashback = sum * 0.05m;
        CashbackAccumulated += cashback;
        TriggerTransactionEvent($"Начислен кэшбэк: {cashback}. Всего кэшбэка: {CashbackAccumulated}");
    }

    public void RedeemCashback()
    {
        if (CashbackAccumulated > 0)
        {
            Balance += CashbackAccumulated;
            TriggerTransactionEvent($"Кэшбэк {CashbackAccumulated} добавлен к балансу.");
            CashbackAccumulated = 0;
        }
        else
        {
            TriggerTransactionEvent("Нет доступного кэшбэка для вывода.");
        }
    }

    public void LoungeVisit()
    {
        if (FreeLoungeVisits > 0)
        {
            FreeLoungeVisits--;
            TriggerTransactionEvent($"Посещение лаунжа использовано. Осталось: {FreeLoungeVisits}");
        }
        else
        {
            TriggerTransactionEvent("Лимит посещений исчерпан.");
        }
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Кэшбэк: {CashbackAccumulated}, Ассистент: {SupportAssistant}, Лаунж визиты: {FreeLoungeVisits}";
    }
}

public class CardHolder<T> where T : CreditCard
{
    public string HolderName { get; set; }
    public T Card { get; set; }

    public CardHolder(string name, T card)
    {
        HolderName = name;
        Card = card;
    }

    public void DisplayCardInfo()
    {
        Console.WriteLine($"Владелец: {HolderName}\n{Card.GetInfo()}");
    }
}

GoldCreditCard goldCard = new GoldCreditCard("1234 5678 9012 3456", "Елена Головач", "31/12/2026", 5000, 10000, 200, 100, 20000);
PremiumCreditCard premiumCard = new PremiumCreditCard("9876 5432 1098 7654", "Максим Комаров", "30/06/2025", 800, 5000, 300, "Личный ассистент", 3);

CardHolder<GoldCreditCard> goldHolder = new CardHolder<GoldCreditCard>("Елена Головач", goldCard);
CardHolder<PremiumCreditCard> premiumHolder = new CardHolder<PremiumCreditCard>("Максим Комаров", premiumCard);

goldHolder.DisplayCardInfo();

goldCard.OnTransaction += (message) => Console.WriteLine($"[GoldCard] {message}");
premiumCard.OnTransaction += (message) => Console.WriteLine($"[PremiumCard] {message}");

goldCard.Pay(1000);
goldCard.AddRewardProgram("Gold Rewards");
goldCard.CheckInsuranceStatus();

Console.WriteLine();
premiumHolder.DisplayCardInfo();
premiumCard.Pay(700);
premiumCard.RedeemCashback();
premiumCard.LoungeVisit();

```
